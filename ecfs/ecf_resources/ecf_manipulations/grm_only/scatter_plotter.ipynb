```python
def entry_check(entry):

    dash = entry.find('/')
    asterisk = entry.find('*')
    numerator = entry[:dash]
    denominator = entry[dash+1:asterisk]
    exponent = entry[asterisk+2:]

    if float(exponent) > 3:
        return 0, 0

    if numerator[1] == 'e':
        nume = int(numerator[2])

    elif numerator[2] == 'e':
        nume = int(numerator[3])

    if denominator[1] == 'e':
        deno = int(denominator[2])

    elif denominator[2] == 'e':
        deno = int(denominator[3])

    return nume, deno
```
